```python
fraud = df[df['Class'] == 1]
nonfraud = df[df['Class'] == 0]
```

```python
# Class separability, using KS test and visual inspection
for col in df.columns[1:29]:  # V1–V28
    plt.figure(figsize=(12,3))
    sns.kdeplot(fraud[col], label='Fraud', fill=True)
    sns.kdeplot(nonfraud[col], label='Non-Fraud', fill=True)
    plt.title(f"Distribution of {col}")
    plt.legend()
    plt.show()
```

```python
from scipy.stats import ks_2samp

for col in df.columns[1:29]:
    stat, p = ks_2samp(fraud[col], nonfraud[col])
    if p < 0.05:  # significant difference
        print(f"{col}: separable (p={p:.2e})")
```

- most columns are separate but we keep the other ones in may be they become meaning full when looking at them in conjunction

```python
# Correlation heatmap
selected_cols = df.columns[1:29]
df_subset = df[selected_cols]
corr_matrix = df_subset.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap of PCA Features")
plt.show()
```

- columns are not corrolated so we don't need to worry about  highly correlated columns.

```python
# plot of time distribution comparing fraud and non-fraud transactions
plt.figure(figsize=(12,6))
sns.histplot(data=df, x="Time", bins=100, color="blue", alpha=0.5, label="Non-Fraud", stat="density")
sns.histplot(data=df[df['Class']==1], x="Time", bins=100, color="red", alpha=0.7, label="Fraud", stat="density")
plt.legend()
plt.title("Transaction Time Distribution")
plt.show()
```

```python
df.drop(columns=['Time'],inplace=True)
```

```python
fraud_amounts = df[df['Class'] == 1]['Amount']
nonfraud_amounts = df[df['Class'] == 0]['Amount']
```

```python
plt.figure(figsize=(12,5))

# Distribution for fraud
sns.kdeplot(fraud_amounts, fill=True, label="Fraud",)

# Distribution for non-fraud (downsample for speed)
sns.kdeplot(nonfraud_amounts.sample(5000), fill=True, label="Non-Fraud")

plt.legend()
plt.title("Transaction Amount Distribution: Fraud vs Non-Fraud")
plt.show()
```

```python
x = df.drop('Class', axis=1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.2, stratify=y, random_state=42)
```

```python
rf = RandomForestClassifier(random_state=1, class_weight='balanced',n_estimators=200,)
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred= rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))
```

```python
print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
```
